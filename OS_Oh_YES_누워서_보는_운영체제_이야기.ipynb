## OS 개론


### Stored program concept
* 폰 노이만 구조, Program(작업)은 실행되기 위해 Main memory 위에 올라와야 한다.
    * Multi Programming : Multi개의 Program이 Main memory 위에 올라와 있는 방식
    * Multi processing system : Multi개의 Processor가 동시에 여러 작업을 parallel로 실행하는 방식
    
* Storage Hierarchy : 폰 노이만 구조의 병목현상(메모리의 값을 읽고 쓰는 구조이기 때문에 기억장치에 걸리는 오랜 시간에 따른 병목 현상)을 해결하기 위해 등장. 가격과 성능 <--> 저장 용량간의 Trade-off 구조.
    * 레지스터 <-> 캐시 <-> 주기억장치 <-> 디스크(SSD <-> HDD)
    * 프로그램이 실행되기 위해서는 반드시 주기억장치(메모리)에 있어야 하며, 메모리에 있는 프로그램의 일부분이 캐시로, 그 중에서 실행되어질 명령어는 처리기 레지스터에 적재되어 실행과정을 밟는다.
    * 캐시나 레지스터는 하드웨어의 제어를 받게되며, 메모리는 관리를 어떻게 하느냐에 따라 시스템의 성능을 좌우하므로 운영체제의 역할이 중요하다.

### OS 
* 사용자 인터페이스와 자원관리를 위한 프로그램의 집합. 다섯가지의 구성요소로 이루어져 있다.
    1. 쉘 : UI 운영체제와 사용자 혹은 운영체제와 프로그램의 의사 소통 역할
    2. 장치 관리 : 키보드, 모니터, Hard disk 등 주변 기기를 관리
    3. 파일 관리 : Directory, 파일 관리
    4. 메모리 관리 : 주기억장치의 관리를 담당. 프로세스 할당 등
    5. 처리기 관리 : CPU의 처리 능력을 각 작업에 효과적으로 배분
* OS 프로그램들은 크게 Kernel과 Utility Program 두 분류로 나뉜다.
    * Kernel : Booting될 때 주 기억 장치에 적재되어 Shutdown 될 때까지 남아있는다. 자주 사용되는 OS Program이 커널에 할당되어 있다.
    * Utility Program : COMMAND.COM, PRINT.EXE 등 UI의 대부분이 여기에 속한다. 필요시에만 주기억장치에 할당되었다가 다시 디스크로 옮겨진다.
* Firmware : 커널중 더 빠른 실행 혹은, 높은 수준의 보호가 필요한 프로그램들을 ROM이나 PLA 같은 칩으로 만든 것

## 프로세스와 스레드
* Process : A program in excution. 프로그램 + 처리기 레지스터 + 스택 데이터, PC 값 등으로 표현되어 커널에 등록된다. (실행을 위해 필요한 환경)
* PCB : Process Control Block. 하나의 프로세스가 생성될 때부터 없어질 때까지 시스템에 어떤 모습으로 존재하는지 정보를 담고 있는 자료구조.
    1. PID(프로세스 고유 번호)
    2. Status(ready, wait 등의 상태)
    3. Priority(우선 순위)
    4. PC(다음에 실행될 명령어의 주소 값)
    5. Memory Pointer(Program과 Data가 저장된 메모리 블록 위치와 공유되는 불록에 대한 포인터)
    6. Context data(문맥 교환시 CPU 레지스터 값 저장)
    7. 할당받은 자원 목록(open 파일 등의 자원 정보)
    8. Accounting Information(CPU를 사용한 시간 등의 정보)
    9. 입출력 정보
* Active state : 프로세스 state가 Ready, Run, Wait 세 가지 상태 중 하나인 경우, process가 메모리 공간의 일부분을 할당받은 상태다. 
    * 이는 다중 프로그래밍의 정도에 영향을 주는 프로세스로, Suspension 대상이 될 수 있다.
* Swapping : 메모리 공간 확보 혹은, 시스템에 위해를 가할 수 있는 수상한 행동을 보이는 경우 메모리에 있는 프로세스를 디스크로 보내거나 다시 불러오는 것.
* Thread의 개념이 추가되면 Process는 부여된 자원의 소유자로, 스레드는 스케줄링의 단위(한 프로세스에 속한 각각의 스레드를 대상으로)가 된다. 
    * 쓰레드는 프로세스가 갖는 자원을 공유하며, 각자 자신의 실행 환경(PC, Stack, Register)을 따로 갖는다. 자신이 속한 process의 상태와 자원은 공유하며, 한 스레드에 의한 공유 데이터 변경시 다른 스레드한테도 영향이 간다. (동기화 요구).
    * 쓰레드도 실행, 준비, 대기와 같은 상태를 가지며, 보류는 없다.
    * 쓰레드 라이브러리가 쓰레드의 스케줄링, 문맥 교환, 생성과 소멸 등을 담당한다.


### CPU 스케줄링 기법
* 시스템의 성능을 높이기 위해 등장한 개념으로, 다음의 경우에 스케줄링이 실행된다.
* 대화형 시스템(Interactive system. 키보드, 아이콘 등을 클릭하여 OS와 특정 행위들을 주고받는 것)에서는 Response time(응답 시간)이, 배치 시스템(Batch system. 일괄 처리)에서는 Throughput(처리량)이 중요하며, 처리량과 응답시간은 서로 반대된다.

* 세 가지 스케줄러
    1. Long-term Scheduler : Job Scheduler라고도 불림. pool에 있는 프로세스들 중에서 어떤 프로세스를 선택해 ready queue로 보낼것인지 결정하는 기법. 새로 작업이 들어오면 먼저 디스크에 놓아뒀다가 장기 스케줄러를 거쳐 프로세스가 되도록 한다. Windows에는 없는 스케줄러.
    2. Mid-term Scheduler : Congestion(Degree of Multiple programming이 높은 경우)이 발생하면, Process를 잠시 하드 디스크로 보내서 대기 상태로 만든다. Multiprogramming Degree를 조절하는 Swapping 기법. 
    3. Short-term Scheduler : CPU Scheduler라고도 불림. ready 상태에 있는 프로세스들 중에서 어느 프로세스에게 CPU를 할당할 것인지 결정하며, 프로세스 스케줄러 또는 Dispatcher(디스패처)에 의해 실행된다. 프로세스 스케줄링 기법을 말한다.

* 프로세스를 실행하기 위해 필요한 세 가지 Queue
    1. Job queue : 메인 메모리에 올라가기 위해 프로그램이 머물러있는 큐
    2. Ready queue : 메인 메모리에 있는 모든 프로세스, run이 되기 위해 ready 상태나 wait 상태로 있다.
    3. Device queue : I/O 디바이스를 위해 기다리는 모든 프로세스

* OS의 스케줄링 전략
    * CPU 스케줄러는 중 단기 스케줄러는 아래의 네 가지 경우에서 스케줄링이 필요하다. 
        1. running state -> waiting state (for I/O) 
        2. running state -> ready state (ex. interrupt 발생, time out)
        3. waiting state -> ready state (ex. done with I/O request)
        4. process terminates
    * 여기서 1번과 4번은 nonpreemptive scheduling이, 2번과 3번은 preemptive scheduling이 필요하다.
    * 이때 단기 스케줄러는 ready queue에 있는 프로세스들 중 어떤 프로세스에 CPU를 할당해줄지 선택하고, 실제 CPU를 할당해주는 것은 Dispatcher로 문맥 교환을 수행한다.
        * Dispatcher의 Latency 대부분은 문맥 교환의 Overhead에서 발생.

### Process 스케줄링
* FCFS (First Come First Served Scheduling) : 비선점. 첫 번째로 온 놈을 먼저 실행시키는 방식.
    * Convoy effect(호위 효과) : 프로세스의 도착 순서에 따라 성능이 매우 차이가 나는 단점이 있다.
   
![img.png](https://user-images.githubusercontent.com/36785390/102321358-ea2cb000-3fc0-11eb-86d1-3ca43861df80.png)
  

*  SJF(Shortest Job First Scheduling) : 비선점. 호위 효과를 해결하기 위해 등장. 가장 짧은 작업시간을(CPU burst) 가진 프로세스를 CPU에 할당시킨다. 
    * 긴 작업시간을 가진 프로세스는 무한 대기(혹은 기아 현상이라고 함) 가능성을 갖고 있다는 단점이 있다. 
        * 이를 해결하기 위해 등장한 것이 Aging 기법. 
    * 각 Process의 남은 작업 시간이 얼마인지 계산해야 하는데, 이를 예측하기가 어렵다는 단점이 있다.
    * 다만, Long term 스케줄러에서는 process time limit을 걸 수 있기 때문에 SJF를 사용함.

  
![img.png](https://user-images.githubusercontent.com/36785390/101979451-8775a480-3ca0-11eb-8f14-cca35a6f6768.png)
  
* 출처 : Silberschatz, Galvin and Gagne의 Operating System Concepts
  

* SRT(Shortest Remaining Time First Scheduling) : 선점. SRTF로도 불림. SJF의 preemption 버전. 현재 실행되는 프로세스보다 짧은 실행 시간을 가진 프로세스가 등장하면 Preemption이 발생해 실행되던 프로세스를 ready state로 만들고, 짧은 시간을 가진 프로세스를 running 시킨다.
    * 이 역시도 starvation이 발생한다.
 
![img.png](https://user-images.githubusercontent.com/36785390/102321363-eac54680-3fc0-11eb-9e3e-2936caa43e02.png)
  
* 출처 : Silberschatz, Galvin and Gagne의 Operating System Concepts
  

* Priority Scheduling : 선점. 우선 순위 번호가 각 프로세스에 부여되어 있다. 낮은 번호일수록 우선순위가 높으며, 가장 높은 우선순위를 가진 프로세스를 실행한다.
    * 기아 현상 발생.
  

* HRRN(Highest Response Ratio Next Scheduling) : 비선점. SJF와 SRT, 우선순위 스케줄링의 starvation 현상을 해결하기 위한 기법. Ready queue에 있는 프로세스들 중에서 Response Ratio(응답률)이 가장 높은 프로세스에게 높은 우선순위를 준다.
    * 응답률 = (대기시간 + CPU Burst) / CPU Burst
    * 대기시간이 길어지며 자연스럽게 응답률이 올라가고, 이를 통해 오래 기다린 프로세스가 우선순위가 높아진다. (Aging 기법) 
  

* Round Robin Scheduling : 선점. 각 프로세스에 부여된 시간 할당량을 기준으로 모든 프로세스를 돌아가며 실행시키는 기법.
    * Time quantum, Time slice라고도 부르는 시간은 어떤 프로세스가 running state에 머물 수 있는 최대 시간이니, 그 시간이 지나면 preempted 당해서 ready queue의 맨 뒤로 이동.
    * Turnaround time은 SJF보다 길지만 Response Time은 더 좋다는 특징이 있다.
  

* Multilevel Queue Scheduling : 선점. 서로 다른 정적 우선순위를 가진 여러 개의 큐를 이용한 스케줄링. 
    * Foreground queue와 Background queue로 나뉘며, Foreground Processes는(interactive processes) RR 알고리즘으로, Background Processes는(batch processes) FCFS 알고리즘으로 처리.
    * 정적(Static) 우선순위, Fiexed priority scheduling 이므로 큐들간에 프로세스의 이동은 불가능하다.
    * Foreground 처리 후 Background를 처리하며, 기아 현상이 발생한다.
  

* MFQ(Multilevel Feedback Queue Scheduling) : 선점. 다단계 큐를 동적 우선순위로 만들어서 프로세스가 큐 사이를 이동할 수 있도록 만들었다.
    * 어떤 프로세스가 CPU 할당을 오래 받았는데도 처리가 끝나지 않았다면, 현재 속한 queue보다 낮은 우선순위의 queue로 이동한다. 또한, 낮은 우선순위 큐에서 너무 오래 대기하는 경우 혹은, I/O가 다 끝나면 높은 우선순위로 이동시키기도 한다.
    * time quantum은 우선순위가 낮아질수록 커지며, FCFS와 RR이 각 큐에서 사용된다.
  

* 실시간(Realtime) 스케줄링 : 모든 프로세스들이 저해진 시간 내에 완료되어야 하는 시스템. RM, EDF 알고리즘이 있으며, 필요한 경우 찾아보자.
  

* Windows의 경우 : 두 개의 Class로 구분하여 우선순위 16 ~ 31은 실시간 클래스를 위해, 0 ~ 15는 일반 클래스를 위해 배정한다. 
    * 실시간 클래스에 속하는 16개의 큐는 정적 우선순위로 운영되므로 다단계 큐로 스케줄링. (각 큐는 RR)
    * 일반 클래스에 속하는 16개의 큐는 동적 우선순위를 위해 MFQ로 구현된다.
    * 입출력 완료시 우선순위를 상향함으로써 입출력 위주 프로세스가 우대받는다. (입출력 위주의 OS이기 때문에)

## 병행 프로세스와 동기화
* Concurrent : 병행. 메모리에 다수의 프로세스가 같이 존재하는 상태.
    * 단일 처리 시스템에선 병행 프로세스들 중 하나를 CPU에 할당
* Parallel : 병렬. 여러 개의 프로세스가 Multi processor에 의해 동시에 처리되는 것
* 동기화(Process Synchronization) : Concurrent process들이 Shared data에 접근하는 경우 공유 자원(Critical Resource. 임계 자원)이 잘못 사용될 위험이 있기 때문에 데이터 일관성을 위해 공유 자원 동시 접근 문제를 해결해야 한다는 개념.
    * Shared data 있는 영역을 Critical Section(임계 영역)이라고 한다. 
    * Shared data에 동시에 접근하는 상황은 Race Condition이라고 한다.
* 동기화 문제를 해결하기 위해선 다음의 세 가지 문제를 해결해야 한다.
    1. Mutual exclusion(MUTEX) : 상호 배재. 임계 영역에 프로세스는 하나만 있어야 한다.
    2. Progress : 임계 영역에서 실행되는 프로세스가 없는 상황에서 그곳으로 들어가려는 프로세스들이 있다면, 누가 들어갈 것인지 결정해야 한다.
        * 잘못하면 서로 양보만 하는 무한 딜레이(Deadlock, 데드락)가 올 수 있다.
    3. Bounded waiting : 프로세스가 임계 영역에 들어가려고 할 때 기아 현상을 예방할 수 있어야 한다. 이는 아래의 해결방법 1을 참고해보자.

### 프로세스 두 개에 대한 동기화 기법 (Peterson)

* 먼저 프로세스가 두 개만 있는 경우다. (SW적 기법)

    * 공유 변수 : int turn = 0;
    * 아래의 두 프로세스를 실행하면 다음과 같은 문제점을 해결하지 못한다. (문제점을 먼저 생각해보고 계속 읽어보자)

```
// process A
while(turn == 0); // do nothing
// critical section
// do something
turn = 1;
```
  
```
// process B

// critical section
// do something
```


#### Peterson 알고리즘 : 오직 두 개의 프로세스만 있는 경우에 사용되는 성공적인 상호배제 알고리즘이다.
* 공유 변수 : int turn = 0; bool flagA = false, flagB = false;
* 이 소스 코드는 Mutex(Mutual exclusion)를 보장하며, whlie 부분에서 Progress 조건을 보장, 두 프로세스 모두 최대 한 번만 기다리며 상대방이 turn을 자기 것으로 만들어주기에 Bounded-waiting 조건까지 만족한다.

```
// process A
// entry section
flagA = true; // 입장할 의사가 있음
turn = 1; // 입장 권한
while(flagB && turn == 1); // do nothing
// critical section
flagA = false; // exit section
```
  
```
// process B
// entry section

```

### 프로세스 N 개에 대한 동기화 기법
* 이번에는 n개의 프로세스 간 상호배제를 위한 기법을 알아보자.

#### Lamport의 Bakery 알고리즘
* 원래 분산 시스템용으로 소개되었으나, n 개의 프로세스들을 대상으로 하는 상호배제 기법으로 사용됨. 은행같이 번호를 오름차순으로 할당한다.
* 프로세스가 받은 번호는 다음에 따라 우선순위가 높은 것으로 결정된다.
    1. 프로세스가 부여받은 번호가 작다. 
    2. 같은 번호를 발급받았을 시, pid가 작은 프로세스가 우선순위가 높다.
* 두 가지 과정을 갖고 있다.
    1. 번호를 뽑는 과정
    2. 번호를 검사하는 과정
* 이 알고리즘 역시 Mutex, Progress, Bounded waiting 조건을 모두 만족한다.
```
    // 공유 변수
    int n = max_process_num;
    bool choosing[n] = {false, };
    int number[n] = {0, };

    // process i의 동기화 과정
    // 1. 번호를 뽑는 과정
    choosing[i] = true;
    number[i] = max(number[0 ~ n-1]) + 1;
    choosing[i] = false;
    // 2. 번호를 검사하는 과정
    for(int j = 0; j < n; j++) {
        while(choosing[j]);
        while(number[j] != 0 && (number[i], i) > (number[j], j));
    }
    // critical section
    number[i] = 0; // exit section
```

* 위에서 소개한 SW 기법들은 운영체제의 특별한 지원 없이, 프로세스 간 협력을 통해 상호 배제를 실현하기 때문에 실행 시 부하가 크고, 오류 가능성도 높다.
* 특히, while(true); 형식으로 인해 CPU를 계속 가동하는 Busy wait(Spinlock, 스핀락)가 발생하기 때문에 CPU를 낭비하는 문제도 발생한다.
 
* 이번에는 하드웨어에서 제공하는 상호배제 기법을 알아보자. 


#### 인터럽트를 금지하는 방법
* 선점 알고리즘 등에 의해 CPU를 뺏길 수 있는 인터럽트를 임계 영역의 실행을 완료할 때까지 발생하지 않도록 한다. (임계 영역에서 인터럽트 차단)
* 이 방법은 시스템의 효율적인 운영을 방해하기 때문에 사용하진 않는다.
```
Interrupt Disable;
// critical section
Interrupt Enable; // exit section
```

#### 하드웨어 명령어를 사용한 기법 1. Test and Set
* 이 기법들은 읽기와 쓰기를 한 명령어 사이클 동안 처리하는 원자성(Atomic)을 가지며, 실행동안 끊기지 않는(Indivisible) 기계적 명령어를 이용한 방법이다.
* testandset 명령어를 이용해서 Mutex 조건과 Progress 조건을 만족한다.
```
boolean testandset(boolean& target) {
    boolean rv = target;
    target = true;
    return rv;
}
```
* 전역변수 lock을 이용해 다음과 같이 사용된다.
```
boolean lock = false;  // lock은 false로 초기화
do {
    while (testandset(lock)) ; //do nothing
    critical section
    lock = false;
    remainder section
} while(true);
```
* 임계영역을 실행준인 프로세스가 있다면 lock의 값이 testandset에 의해 true로 바뀌고, 이후 진입을 시도하는 프로세스들은 바쁘게 실행을 대기하게 된다.
* 다만, 프로세스 도착 순서와 무관하게 critical section에 진입하기 때문에 fair 하지 않다. (Bounded waiting 조건 불만족)
* Bounded waiting을 만족시키기 위해 다음과 같이 바꿔보자.
```
bool lock = false;
bool waiting[n] = {false, }; // true : critical section 진입하고 싶다
do {
    bool key = true; // testandset 결과 값 받아줄 용도
    waiting[i] = true;
    while(waiting[i] && key) 
        key = testandset(lock);
    waiting[i] = false;

    // critical section

    // exit section : 기다리고 있는 process가 있는 경우와 없는 경우를 나눈다.
    // 우선 waiting중인 j 프로세스를 찾는다.
    int j = (i + 1) % n;
    while(j != i && !waiting[j])
        j = (i + 1) % n;
    //만약 j == i면 waiting 중인 프로세스가 없다.
    if(j == i) 
        lock = false; // 초기화
    else
        waiting[j] = false; // j번째 프로세스 실행하렴
    //remainder section
} while(1);
```
* 최대 n-1번만 기다리면 됨. 만약 이해가 안가면 [링크](https://blog.naver.com/and_lamyland/221188862284)

#### 하드웨어 명령어를 사용한 기법 2. Compare and Swap (exchange)
* 고전적인 swap 함수를 사용하거나 아래의 함수를 사용한다
```
int compare_and_swap(int *value, int expected, int new_value) {
    int temp = *value;

    if (*vale == expected) 
        *value = new_value;
 
    return temp;
}
```
* compare_and_swap 함수는 다음과 같이 사용한다.
```
int lock = 0;
do {
	while (compare_and_swap(&lock, 0, 1) != 0);
	// critical section
	lock = 0; // exit section
	// remainder section
} while (true);
``` 

#### 소프트웨어 tool을 이용한 기법 1. Mutex Lock
* 위의 하드웨어 차원에서 제공하는 동기화 기법은 응용 프로그래머 입장에서 사용하기 어렵기 때문에 소프트웨어 툴이 만들어졌다.
* 이 알고리즘은 acquire()와 release() 함수를 사용하며, bool available을 사용해 lock이 잠겨있는지, 풀려있는지 결정한다. 다음의 함수와 사용 예를 살펴보자.
```
acquire() {
	while (!available); //do nothing == busy waiting == spinlock == waiting in running state
	available = false;
}
release() {
	available = true;
}
```
```
do {
	acquire lock

	critical section

	release lock

	remainder section
} while (true);
```
* 여기서 중요한 것은 acquire()와 release()는 Atomic, Indivisible 한 함수여야 한다.

#### 소프트웨어 tool을 이용한 기법 2. Semaphore
* Dijkstra가 제안한 세마포어는 동기화 문제를 해결할 뿐 아니라 Spinlock을 막기 위해 busy waiting하는 프로세스를 wait state로 보냅니다.
* 세마포어는 세마포어 구조체, wait(), signal()로 구성된다.
    * wait == down == P. P = 네덜란드에서 wait의 이니셜
    * signal == up == V. V = 네덜란드에서 signal의 이니셜
* 세마포어는 Binary(이진) 세마포어와, Counting(계수, Integer 정수) 세마포어로 나뉜다.
```
struct semaphore {
	int value;
	semaphore *list; //wait list. queue
};
```
* 이진 세마포어 먼저 살펴보자. (Mutex lock과 동일)
```
wait(semaphore *S) {
	if (S->value == 0) {
		add this process to S->list; //block 시키기 위해 (wait state로 보내기 위해)
		block(); //못 들어가니 재워
	}
	else S->value = 0;
}
signal(semaphore *S) {
	if (not empty S->list) {
		remove a process P from S->list; //place P onto ready state
		wakeup(P); //당신의 차례니 깨워
	}
	else S->value = 1;
}
```
```
do {
	wait(mutex);

	critical section

	signal(mutex);

	remainder section
} while (true);
```
* 이진 세마포어에서 value는 1로 초기화되어 있다.

* Counting 세마포어
```
wait(semaphore *S) {
	S->value--;
	if (S->value < 0) {
		add this process to S->list; //block 시키기 위해 (wait state로 보내기 위해)
		block(); //못 들어가니 재워
	}
}
signal(semaphore *S) {
	S->value++;
	if (S->value <= 0) {
		remove a process P from S->list; //place P onto ready state
		wakeup(P); //당신의 차례니 깨워
	}
}
```
* value는 n 또는 0으로 초기화된다. (즉, 1이 아니다. 따라서 이진 세마포어와 계수 세마포어는 value 초기 값으로 쉽게 구분된다)
* 세마포어에서는 spinlock 문제를 해결하며 CPU의 활용도를 높이려 하고 있다. 다만, 다음과 같은 경우에는 오히려 spinlock이 있는 것이 더 최적화된 상태일수도 있다.
    * Critical section이 매우 짧고, 진입하고자 하는 프로세스가 많지 않아 busy waiting 할 시간이 적으면, busy waiting 하는 구조가 block & wake up보다 유리
        * process state를 변경하기 위해 드는 overhead가 없기 때문에
    * 따라서 상황에 맞게 동기화 기법을 이용하는 것이 현명한 방법.

### 프로세스 동기화 문제(Producer Consumer, Dining Philosophers Problem)와 모니터(Monitor)

#### 생산자 - 소비자 문제
* 생산자는 데이터를 만들어 원형 버퍼에 채우고, 소비자는 버퍼에 있는 데이터를 꺼내 소비하는 프로세스다.
    * 버퍼는 n개의 공간을 가진 공유 자원이므로, 상호배제 되어야 한다.
* 이 문제를 풀기 위해 필요한 자료는 다음과 같다.
    ```
    int n;
    semaphore mutex = 1; 
    semaphore empty = n;
    semaphore full = 0;
    ```

* producer는 아래와 같이 설계한다.
    ```
    do {
        // Producer process가 item을 생성
        wait(empty);
        wait(mutex);
        // 버퍼에 새로 만들어진 item을 추가
        signal(mutex);
        signal(full);	
    } while (true);
    ```

* consumer는 아래와 같이 설계한다. 
    ```
    do {
        wait(full);
        wait(mutex);
        // 버퍼에서 item을 빼옴
        signal(mutex);
        signal(empty);
        // Consumer process가 item을 사용
    } while (true);
    ```

* 실행 예제는 다음과 같다.
    1. Producer가 item을 만든다.
    2. 임계 영역에 진입하며 wait(empty), wait(mutex)를 호출한다.
        * wait(empty) : n--;
            empty의 초깃값 n : 버퍼에 n개의 공간이 비어있다.
        * wait(mutex) : 상호 배제 조건을 만족시키기 위한 함수. 임계영역 진입.
    3. 버퍼에 아이템을 채운다.
    4. 임계 영역을 빠져나오며 signal(mutex), signal(full)을 호출한다.
        * signal(mutex) : 임계 영역 빠져나옴
        * signal(full) : 0인 초기 값을 1로 만들어 줌.
            * full의 초깃값 0 : 버퍼에 공간이 0 만큼 채워져 있다.
    5. Consumer가 item을 빼오고자 wait(full), wait(muxte)를 호출한다.
        * wait(full) : 버퍼에 아이템이 몇 개만큼 들어있는지 확인. 0개면 못빼옴
        * wait(mutex) : 임계 영역 진입
    6. 버퍼에서 아이템을 빼온 뒤, signal(mutex)로 임계 영역 빠져나오고, signal(empty)로 비어있는 공간의 수를 1 늘려준다.
  

#### 여러 개의 세마포어를 사용할 때 주의할 점
* 세마포어 사용하는 명령어 위치를 세심하게 따져봐야 한다.
* 예를 들어, comsumer가 wait(full) -> wait(mutex)하는 구조에서 wait(mutex) -> wait(full)하는 구조로 바뀌게 된다고 가정해보자.
    * 소비자가 wait(mutex)해서 임계영역에 아무도 못오게 선언했는데, 만약 wait(full) 때문에 block되면 deadlock에 걸려버린다.
        

#### 식사하는 철학자 문제
 
 
![img.png](https://user-images.githubusercontent.com/36785390/102321353-e8fb8300-3fc0-11eb-91e5-b628c30d0582.png)
 
 
* 출처 : Silberschatz, Galvin and Gagne의 Operating System Concepts

* 철학자 다섯 명이 원형 테이블에서 식사를 하려고 한다.
    1. 테이블 중앙에 음식을 먹으려고 할 때, 철학자 왼쪽과 오른쪽에 있는 젓가락을 잡으려고 시도한다.
    2. 두 젓가락을 모두 잡으면 식사가 가능하다.
    3. 젓가락이 사용중이라면 사용할 수 있을 때까지 기다린다.
    4. 식사가 끝나면 젓가락을 다시 테이블에 올려둔다.

* 이 문제를 해결하기 위해 이진 세마포어 다섯 개를 사용한다.
    ```
    semaphore chopstick[5]; // binary semaphore
    do {
        wait(chopstick[(i + 4) % 5]);	//왼쪽의 젓가락을 집자.
        wait(chopstick[(i + 1) % 5]); // 오른쪽의 젓가락을 집자.
        //젓가락 둘 다 집으면 통과
        Eating;
        signal(chopstick[(i + 4) % 5]);	//왼쪽 내려놔.
        signal(chopstick[(i + 1) % 5]);	// 오른쪽 내려놔.
        Thinking;
    } while (true);
    ```

* 얼핏 단순해보이는 이 동기화 방법에는 데드락이 발생하는 치명적인 문제가 있다.
    * 0번 철학자가 wait(chopstick[(i + 4) % 5])를 호출 -> 4번 철학자가 wait(chopstick[(i + 4) % 5])를 호출 -> 3번 철학자가 wait(chopstick[(i + 4) % 5])를 호출 -> 2번 철학자가 wait(chopstick[(i + 4) % 5])를 호출 -> 1번 철학자가 wait(chopstick[(i + 4) % 5])를 호출 -> 데드락
    * 이 문제를 해결하라면 식탁에 한 번에 4 명만 앉도록 하거나, 홀수 번호의 철학자는 왼쪽 포크부터, 짝수 번호의 철학자는 오른쪽 포크부터 집게하면 된다.

* 이런 잘못된 설계에 의해 세마포어는 데드락을 발생시킬 수 있는 치명적인 단점이 있다.
* 따라서 지금부터는 강력한 동기화 도구인 Monitor를 알아야한다.

#### Monitors (모니터) 
* 모니터는 ADT(Abstract Data Type)의 안전한 High Level Synchronization Software Construct다.
* 모니터는 프로그래머가 정의한 함수에 MUTEX를 제공하며, 모니터 안에선 항상 하나의 프로세스만 running할 수 있다.
* 모니터 내의 프로세스가 실행 중 특정 조건 때문에 대기해야하는 경우(I/O Interrupt 같은), 모니터를 양보하기 위해 자신은 Waiting Signaler Queue(신호자 대기 큐)에 들어가게 된다.
 

![img.png](https://user-images.githubusercontent.com/36785390/102322740-e0a44780-3fc2-11eb-9c01-5086092d2756.png)
 

* condition variables는 신호자 대기 큐를 의미하며, 대기가 필요한 프로세스를 wait(x), wait(y) 과정에서 처리해주게 설계한다.
    * 생산자-소비자 문제를 예로 들자.
        * 생산자 A가 item을 생산하고 append하는 과정에서 버퍼 공간이 부족해 대기가 필요하다면 not empty condition waiting queue에 넣어준다.
        * 소비자 B가 item을 소비해서 공간이 하나 비면 signal(empty)를 호출하며 empty condition waiting queue에 있는 A를 살려준다.
    * 다른 예로, I/O Interrupt로 인해 대기 큐에 들어가게 된다면, 다음과 같은 상황이 연출된다.
        * 프로세스 P가 queue에서 wait중인데, 
        * 프로세스 Q가 signal을 통해 프로세스 P를 깨우면 다음과 같은 두 가지 선택지가 생김
            1. Signal and Wait : 살려준 프로세스 Q가 모니터를 떠날 때까지 프로세스 P는 실행을 연기
            2. Signal and continue : 살아난 프로세스 P가 Q보다 먼저 실행을 함.
        * 보통은 Signal and wait 방식을 사용.

#### 모니터를 이용한 식사하는 철학자 문제 풀이
* 철학자는 앞으로 세 가지 상태를 갖는다.
    * Thinking, Hungry, Eating
        * Thinking : 식사할 의지 없음
        * Hungry : 식사를 하고자 함
        * Eating : 식사 중
    ```
    monitor DiningPhilosophers {
            enum {THINKING, HUNGRY, EATING} state[5];
            condition self[5];
            void pickup(int i) {
                state[i] = HUNGRY;
                test(i);
                if (state[i] != EATING)
                    self[i].wait();
            }
            void putdown(int i) {
                state[i] = THINKING;
                test((i + 4) % 5);
                test((i + 1) % 5);
            }
            void test(int i) {
                if ((state[(i + 4) % 5] != EATING) &&
                    (state[i] == HUNGRY) &&
                    (state[(i + 1) % 5] != EATING)) {
                    state[i] = EATING;
                    self[i].signal();
                }
            }
            initialization_code() {
                for (int i = 0; i < 5; i++)
                    state[i] = THINKING;
            }
    }
    ```

* 실행 순서는 다음과 같다.
    ```
    // sequence : 
    DiningPhilosophers.pickup(i);
    // Eating
    DiningPhilosophers.putdown(i);
    ```

* 고민해볼점
    1. 맨 처음에 식사하는 철학자에게 test(i)의 self[i].signal() 호출은 아무 의미가 없다. 언제 의미가 있는 코드일까?
        * 옆에 기다리고 있는 사람이 있을 때부터. 즉, 맨 처음에는 의미가 없음 (아무도 안 기다림)
    2. Starvation (기아 현상)이 발생한다. 언제 발생할까?
        * 계속 밥을 먹으려는 두 사람 사이에 한 명이 wait하고 있는 상태

#### Spooler와 De-spooler
* Spooling(스풀링)은 원형 버퍼를 관리하는 모니터를 이용해 구현된다.
    * 스풀러 프로세스는 디스크와 같은 보조기억 장치에 빠른 속도로 출력을 하는 프로세스고, 디스풀러 프로세스는 디스크에 임시로 출력된 내용을 실제로 프린터로 출력하는 시스템 프로세스다.
    * 상대적으로 속도가 느린 프린터와 직접적으로 관련된 작업을 하는 디스풀러 프로세스는 스풀러 프로세스에 비해서 처리 속도가 느리다.
    * 이 두 프로세스 간의 속도 차이를 완화해주기 위해 스풀러 프로세스는 생산자, 디스풀러 프로세스는 소비자가 되어 원형 버퍼를 공유한다.

## 교착 상태 (Deadlock)
* 둘 이상의 프로세스가 각자가 갖고 있는 자원을 보유한 채로 외부적 조치가 없는한 영원히 멈춰있는 상태.
    * 자원(Resource) 
        1. 선점 가능성에 따른 자원의 분류
            1. Preemptible Resource(선점 가능 자원) : CPU나 메모리와 같은 자원처럼 한 프로세스에 의해 사용되는 도중 선점(Preemption)되어 다른 프로세스에게 할당(Allocation)해 주었다가 이 후 다시 원래의 프로세스에게 돌려주어도 되는 자원들
            2. Nonpreemptible Resource(선점 불가능 자원) : 선점이 될 경우 자원을 뺏긴 프로세스는 정상적인 진행을 포기해야 하는 불이익을 받게되는 경우.
                * 프린터 출력 중 섞이면 안돼
        2. 사용되는 방식에 따른 자원의 분류
            1. Sharable Resource(공유 가능 자원) : 한 프로세스에게 할당된 자원을 다른 프로세스가 할당받아 같이 사용할 수 있는 자원. 공유 가능한 프로그램(시스템 프로그램이나 사용자 프로그램 등)과 공유 데이터 등이 있다.
            2. Exclusive Resource(배타적 사용 자원) : CPU, 메모리, 테으프, 모니터 등이 있으며, 자신이 사용중일 땐, 남이 못쓰게 한다.
        3. 자원의 재사용성에 따른 분류
            1. Serially Reusable Resource(순차적 재사용 가능 자원) : 프로세스가 실행 후 반납한 자원이 해제되지 않고 다른 프로세스도 사용 가능한 자원. CPU, 메모리, 하드디스크 등.
            2. Consumable Resource(소모성 자원) : 일시적으로 생성되었다가 사용후 없어지는 자원. 시그널이나 메시지 등.

* 교착 상태의 원인은 다음과 같다.
    1. Mutual Exclusion : 상호배제, 자원의 배타적 사용, 시스템이 보유한 한정적이고 공유 불가능한(배타적) 자원에 대해 프로세스들의 사용 경쟁이 발생할 때 문제 발생
    2. Hold and Wait : 점유 대기, 프로세스가 실행에 필요한 자원을 갖고있는 상태에서 추가로 다른 자원을 필요로 했다가 대기 상태가 되는 경우.
    3. No Preemption : 비선점, 선점의 권한이 주어진다면 프로세스들은 자원 때문에 대기할 필요가 없으므로 교착 상태는 발생하지 않는다.
    4. Circular Wait : 환형 대기, 각 프로세스는 순환적으로 다음 프로세스가 요구하는 자원을 가지고 있다. (그래프에 사이클이 발생하는 경우)


### 교착 상태 해결 방법
1. Prevention (예방) : 교착 상태의 발생 자체를 막도록 한 방법
    1. Mutual Exclusion 조건 배제 : 모든 자원을 공유 가능 자원으로 하면 됨. (절대 불가능)
        * 당장 프린터만 공유하더라도 난리가 날거임.
    2. 자원의 부분 할당을 배제 : 프로세스 실행에 필요한 자원을 모두 할당하는 방법으로, Hold and Wait을 제거한다. 다음과 같은 이유 등으로 이 방식도 사용 못한다.
        * 필요한 자원이 확보되지 못한 다른 프로세스들은 대기에 빠진다. 
        * 프로세스가 실행되는 순간에 자신에게 필요한 자원이 전부 필요한 경우는 거의 없다.
    3. Preemption : 모든 자원이 선점 가능하도록 하는 경우, 실행 도중 자원을 반납시킨 프로세스는 비정상적인 종료를 할 가능성이 있으며, 처음부터 다시 계산을 실행해야 하는 경우가 발생할 수 있다. 무한 대기도 발생할 수 있으니 사용 못한다.
    4. 환형 대기 배제 : 시스템 측에서 자원 할당 순서를 매김. 자원의 사용 및 할당 순서를 정하고 프로세스들은 이 순서를 지키면 교착 상태를 없앨 수 있다만, 모든 프로세스와 자원들에 대해 이 방식을 설정하기는 사실상 힘들다.
        * 실제로 필요한 자원을 확보하기 위해 지금 당장 필요 없는 자원을 확보하고, 필요한 자원 사용하기 위해 대기하고 ...

2. Avoidance (회피) : 데드락의 발생 가능성을 계속 검사해서 데드락 발생 가능성이 있다면 회피해 버리는 방식. 예방과 회피 모두 교착 상태가 발생하지 않도록 함.
    * Safe state : 데드락 가능성이 없는 프로세스 상태
    * Unsafe state : safe state인 프로세스가 unsafe state가 된다면, 데드락이 발생한다는 뜻.
    * 회피 기법은 Safe state에 있던 프로세스가 Unsafe state로 간다면, state를 바꾸게 한 행동을 찾아 행동을 취소하게 한다든지의 방식으로 다시 safe state로 돌아가게 만들어야 함.
    * Safe sequence : 프로세스들이 현재 할당받은 자원과, 더 할당받고 싶어하는 자원이 뭔지 확인하고, 어떤 순서로 자원을 할당했을 시 데드락이 발생하지 않고 모든 프로세스의 요구를 처리해줄 수 있는 순서를 말함.
    * 세이프 시퀀스가 하나라도 있으면 데드락은 회피가 가능하다.

3. Detection & Recovery (탐지 & 회복, 복구) : 탐지와 복구(회복)는 하나로 묶어서 사용한다. 알고리즘을 통해 현재 시스템에 데드락이 있는지 찾고 복구함.
    * 탐지 기법 : 교착 상태를 탐지하기 위해 자원 할당 그래프(RAG, Resourece Allocation Graph)를 사용한다.
        * 자원이 Single instance인 경우 : Wait for graph로 바꿔서 Cycle이 있는지 확인
        * 자원이 Multiple instance인 경우 : 회피 기법의 Safety Algorithm과 비슷한 방법으로 데드락 프로세스를 찾는다. (위의 링크에서 자세한 내용 확인하자)
    * 회복 기법 : 데드락 회복 방법. 프로세스 종료 방법과 자원 선점에 의한 방식이 있다.
        1. Process Termination : Abort one or more Processes. 강제로 종료된 프로세스들은 희생이 요구되므로, 
            1. 종료 비용(Termination Cost, 프로세스가 잃게 되는 일의 양)을 따져서 Lowest-termination-cost Process 먼저 종료시킨다.
            2. 교착 상태에 있는 프로세스들의 집합에서 가능한 모든 부분 집합을 만들어, 최소의 비용이 드는 부분 집합에 속하는 프로세스들을 한 번에 종료시킨다.
        2. Resource Preemption : Preempt some resources. 
            1. Selecting a victim : 어느 프로세스의 어떤 자원을 뺏을지 결정함.
            2. Rollback : 자원이 뺏긴 프로세스를 safe state로 rollback 하고 나중에 다시 restart하거나 처음부터 다시하거나 결정함.
            3. Starvation : 우선순위의 기준에 따라 희생하는 놈이 계속 희생을 하기 때문에 Starvation이 발생할 수 있음.

4. Ignore (무시) : 예전 시스템은 해당 프로세스를 kill하거나 재부팅함. 다만 최근에는 병렬, 분산 처리가 일상이 되어 데드락을 무시하기는 힘듦.

## 메모리 관리
* 프로그램이 실행되기 위해서는 메모리에 올라와 있어야 함.
    * 당연히도 메모리 관리를 잘 하면 CPU의 효율적인 사용과 사용자에게 빠른 응답성을 줄 수 있겠지.


* 이번 장에서는 Contiguous, 연속적으로 프로그램을 메모리에 할당하는 방법에 대해서 다룰 것임.

* 메모리 구성과 관련해서 염두해야 할 것 
    1. Multiprogramming Degree 
        * 다중 프로그래밍 정도는 한 번에 여러 개의 프로그램이 같이 실행될지 아닐지 등을 결정하는 용도로 사용.
        * Degree는 메모리에 있는 프로세스 개수를 말함.
        * 정도가 n일 경우 각 프로세스에 얼만큼의 메모리를 줄 것인지 고민해야 하고, 이는 메모리 분할(Partition)과 연계된다.
            * 분할에는 Fixed(Static, 고정, 정적), Variable(Dynamic, 가변, 동적) 두 방식이 있겠지.

    2. 연속적으로 할당할 것인가, 비연속적으로 할당할 것인가.


* 메모리 관리를 위해 다음과 같은 기법들이 존재한다.
    1. 반입 기법 (적재 기법, Fetch strategy)
        * 프로세스에게 언제 메모리를 할당해줄 것인가 다루는 기법
            1. Demand 요구 : 대부분의 시스템에서 채택. 적재 요구가 있을 때 적재.
            2. Anticipatory 예상 : 적재 요구가 있을 것으로 예상하고 미리 적재해두는 방법

    2. 배치 기법 (Placement)
        * 메모리 어느 공간에 프로세스를 적재할 것인가.
        * 최초 적합 First fit : free 리스트의 첫 노드부터 시작하여 제일 먼저 발견되는 프로세스 수용 가능한 크기의 공간에 할당
        * 최적 적합 Best fit : free 리스트를 끝까지 탐색하여 요구되는 크기보다 더 크되 그 차이가 가장 적은 노드(공간)에 할당
        * 최악 적합 Worst fit : free 리스트를 탐색하며 차이가 제일 많이 나는 공간에 할당
        * 최적 적합은 할당 후 남는 빈 공간이 가장 적긴 하지만, 이런 빈 공간들이 이후의 적재에 거의 활용되지 못한다는 단점이 있다. (Hole)
            * 또한, 모든 리스트를 탐색해야 해서 시간적 부담이 크다.
        * 대부분의 경우 최초 적합이 제일 빠르지만, 시간이 지날수록 free 리스트의 앞부분부터 매우 작은 노드들(대부분 Hole)이 생길 것이기에 리스트 탐색 시간이 길어질 것이다.
        * next fit : 최초 적합의 단점을 해결하고자 free 리스트를 순환 구조로(Circular) 만든 다음 할당이 가능한 노드가 선택될 때마다 헤더 포인터를 다음 노드로 옮기게 한다.

    3. 할당 기법 (Allocation)
        * 프로세스에게 메모리 공간을 얼마나 할당해줄 것인가.

    4. 교체 기법 (Replacement)
        * 메모리 공간이 부족할 경우 새로 적재돼야 할 프로세스를 위해 이미 메모리에 있는 프로세스 중 어떤 것을 골라 디스크로 보낼 것인가.


* 가변 분할에서의 다중 프로그래밍
    * 가변 분할이란 분할의 시기와 개수 그리고 크기가 사전에 정해진 바 없이, 프로세스를 수용할 때 그 크기만큼 메모리 공간을 할당해 줌을 말한다.
    * 이를 통해 내부 단편화(Internal Fragmentation, 분할된 메모리 안에 프로세스를 꽉 채우지 못해서 공간이 남음) 현상을 방지할 수 있지만, 관리가 복잡해져 오버헤드가 커진다.
    * 또한, 외부 단편화(External Fragmentation, 프로세스를 할당받을 수 있는 메모리 크기가 너무 작아서 수용하지 못하기 때문에 생기는 잉여 공간)도 발생할 수 있다.

    * 이 방식에선 free space에 대한 연결 리스트와 used space에 대한 연길 리스트를 이용한다.
        * ex) 크기가 10인 Process A, 20인 Process B 실행 예정
                초기 메모리 Total Free Space : 100
                1. 프로세스 A 적재 후 FS(Free Space) -> 90
                                      US(Used Space) -> 10
                2. 프로세스 B 적재 후 FS -> 70
                                      US -> 10 -> 20
                3. 프로세스 A 메모리 반납 후 FS -> 70 -> 10
                                             US -> 20
    
    * 만약 위의 예에서 새로운 10짜리 프로세스 C를 적재해야 한다면 FS의 70과 10중 어디로 적재하는게 좋을까? 
        * 이때 나온게 배치 기법. 최초, 최적, 최악


* 50% 규칙 50-percent Rule
    * 실제로 빈 공간이 많음에도, 대부분이 홀이어서 별로 크지 않은 프로세스의 적재 요구조차 수용할 수 없는 경우가 많기 때문에 생긴 용어.
    * 실제로는 메모리의 3분의 1이 홀이 된다는 실험 결과가 있음.
    * 결국, 작은 빈 공간을 합쳐 더 큰 빈 공간을 만들어야 하는데, 이를 위해 다음과 같이 두 가지 분류가생긴다.

    1. 인접한(Adjacent) 빈 공간의 병합(Coalescing)
        * 빈 공간으로 반납될 때 인접한 빈 공간이 있다면 이들을 합친다.
        * 프로세스가 메모리 반납할 때마다 실행된다.

    2. 빈 공간 전부 통합 (Compaction)
        * 사용중인 공간들을 메모리의 한쪽으로 밀착시키고, 흩어져 있던 빈 공간들을 전부 합쳐 하나의 큰 빈 공간으로 만든다.
            * 다만, 프로세스들이 주소 재배치를 해야 하므로 시간이 매우 오래 걸림.
    

* 버디(Buddy) 시스템
    * 위의 두 공간 관리 기법의 절충안
    * 가변 분할과 마찬가지로 최초에 큰 빈 공간 하나로 시작한다.
    * 프로세스 적재 요구가 있을 때, 메모리는 요구한 크기보다 크되 차이가 가장 작게 나는 2의 승수 크기로 분할되어 할당시킨다.
        * 이때 같은 크기로 분할된 인접 공간을 버디라 부른다.
    * 반납되는 빈 공간은 분할 시 정해졌던 자신의 버디가 빈 공간일 때 병합되어 크기를 2의 배수로 늘려간다.

    * [예제 링크](http://www.jidum.com/jidums/view.do?jidumId=484)

    * 누가 나의 버디인 지를 알아야 병합이 제대로 이뤄진다. 버디를 찾는 계산은 다음과 같다.
        * 크기가 2^k, 메모리 시작 주소가 x인 공간이 있을 때, 이 공간의 버디의 시작 주소는 x mod 2^(k+1) == 0 ? (x + 2^k) : (x - 2^k)

    * 미리 고정된 분할도 아니지만, 프로세스의 크기만큼 할당하는 가변 분할도 아니지만 고정 분할보다 내부 단편화 사정이 좋다.

## 가상 메모리 (Virtual Memory)
* 가상 메모리는 사용자들의 프로그램이 아무리 크더라도 프로그램의 일부만 적재하면서 엄청나게 큰 메모리가 있는 것처럼 프로세스를 관리하는 방식이다.
    * 가상 메모리에서 가장 중요한 것은 Address 관리와 Mapping이다.
        * 실행 중인 프로그램에서 참조하는 주소가 실제 메모리에 있는 주소와 달라서, 메모리상의 주소로 변환이 필요로 할 때 하는 것이 Mapping, 사상.
        * 실주소는 셀제 메모리상의 주소를 의미.


* 먼저 모든 프로그램은 작은 조각들로 나눠지게 되는데, 조각들의 크기를 모두 같도록 하면 Page, 다르게 하면 Segment라고 부른다.
    * Page와 Segment는 메모리와 디스크 사이에서 한 번에 전송되는 전송 단위가 된다.(Block)


### Paging 페이징
* 모든 프로세스들이 같은 크기의 조각들로 나뉘며, 이 단위를 페이지라 부른다.
    * 메모리 역시 Frame(프레임)이라 불리는 페이지와 같은 크기로 나눠지며, 일련의 번호가 매겨진다.
        * 디스크와 메모리를 오가며 교체되는 단위
    * 페이지의 크기는 일반적으로 수 KByte 정도이며, 2의 승수 배 크기로 잡는다.


* 한 프로세스의 전체 페이지들은 디스크에 저장되고, 이 중 몇 개가 메모리에 비연속적으로 다른 프로세스들의 페이지들과 섞여서 메모리에 적재된다.


* Page Table
    * 운영체제는 가상 주소를 실주소로 변환하기 위해 페이지 사상 테이블(Map Table)이라는 것을 운영하며, 이 테이블의 크기는 해당 프로세스의 페이지 개수에 비례한다.
        * k 개의 페이지를 갖는 프로세스의 페이지 테이블은 k 개의 엔트리(Entry)로 구성되고, 엔트리 하나의 크기는 보통 4 kbyte
    * Entry에 있는 정보
        * 페이지가 메모리에 적재되어 있는가를 나타내는 Residence bit (존재 비트)
            * 1인 경우 적재 되어 있는 프레임의 번호가 저장됨
            * 0인 경우 이 페이지가 저장되어 있는 디스크의 주소를 나타내는 Field 들이 저장됨
        * 페이지 번호(p)와 페이지 내에서의 위치(d)
    
    * 페이지 테이블은 메모리의 커널 영역에 보관됨.
    * 실행 중인 프로세스의 페이지 테이블 시작 주소는 Page Table Origin Register(페이지 테이블 기준 레지스터)에 들어있음.

    * 실행 과정
        1. 기준 레지스터 값에 페이지 p의 엔트리를 찾은 후 존재 비트를 확인한다.
        2. 1일 경우 p가 적재되어 있는 프레임 번호(f)를 알 수 있으므로, 이 번호 값에 페에지 크기를 곱하면 메모리에서 이 프레임의 시작 주소를 얻게 되고, d를 더하면 실주소를 접근할 수 있게 된다.
        3. 0일 경우 페이지가 메모리에 없기 때문에 디스크 주소로부터 먼저 이 페이지를 메모리에 적재한다.
            * 이후, 그 다음 이 엔트리의 존재 비트를 1로 바꾸고 적재된 프레임 번호를 기입한 후 mapping을 다시 진행한다.
        
    * 단점
        1. 페이지 테이블 자체도 크기가 매우 커질 수 있기 때문에 다중 프로그래밍 정도에 포함되는 프로세스들의 테이블들만 보관하고, 나머지는 디스크에 둠.
        2. 두 번의 메모리 접근이 요구 됨. (가상 주소 사상을 위해 페에지 테이블 접근 후 실주소 접근)
            * 이를 줄이는게 가상 메모리의 단점을 줄이는 지름길임
            * 기본적으로 사상이 요구될 가능성이 높은 테이블들을 메모리에 올려놔야 하며, 연관 메모리(Associative Memory)를 쓰기도 한다.
                * 주소가 아닌 내용(Content)에 의해 접근하는 메모리 장치
                * 이 방식은 데이터 내용을 병렬 탐색 하기에 적합함
                * 탐색은 전체 워드 또는 한 워드 내의 일부를 갖고 실행할 수 있음
    

* TLB(Translation Lookaside Buffer)의 사용
    * TLB는 고속 캐시의 일종으로, 주소가 아닌 Key 값으로 찾고자 하는 워드를 동시에 접근하는 연관 메모리다.
    * 페이지 테이블을 메모리에 두고 접근할 때의 느림을 극복하기 위해 사용함.
    * 최근에 빈번하게 검색된 엔트리들을 TLB에 넣되, 페이지 번호(p)를 키 값으로 동시 검색을 하므로, TLB에 저장되는 각 엔트리는 페이지 번호도 함께 표시되어야 한다.
    
    * Mapping 과정
        1. 가상주소 p를 Key 값으로 TLB 검사
        2. 페이지 번호를 갖고 있는 엔트리가 있을 경우, 그 엔트리에 적혀있는 프레임 번호를 갖고 실주소에 접근 (이하 위에서 서술한 과정 진행)
            * TLB는 존재 비트를 확인할 필요가 없음.
        3. TLB에서 p를 갖는 엔트리 검색에 실패하면 메모리에 있는 페이지 테이블로부터 Mapping이 진행되어 실주소에 접근 (이하 위에서 서술한 과정 진행)
        4. 이후, 이 엔트리는 TLB에 추가되며 페이지 교체가 일어날 수 있다.
    
    * TLB의 성능을 끌어올리기 위해선 Hit ratio(적중률)가 높아야 하는데, 보통 90% 이상은 나오도록 만든다.


* 페이지의 보호와 공유 (Protection and Sharing)
    * 보호 : 페이지에 대해 읽기나 쓰기 작업을 어떻게 제한할 것이며, 다른 프로세스의 주소 공간으로 침범하지 못하도록 어떻게 할 것인가?
        * 페이지 테이블의 각 엔트리에 해당 페이지에 대한 Protection Bits(보호 비트)를 두어 허용되는 접근을 설정할 수 있다.
        * 프레임의 시작 주소에 더해지는 offset 값이 페이지의 크기를 넘게 되면 다른 프레임으로의 접근이 가능해지기 때문에 트랩을 통한 보호가 이뤄져야 한다.

    * 공유 : 프로그램은 여러 개의 프로세스를 만들 수 있다. 다수의 사용자가 하나의 프로그램으로 다수의 프로세스를 실행한다면, 중복되는 모든 데이터를 메모리에 올려야하는 불상사가 생길 수 있다.
        * 따라서 중복되는 데이터는 하나만 메모리에 올려 공유할 수 있도록 한다.
        * 다만, 사용자는 각자의 실행 위치가 다른 동시에 수정되는 각자의 데이터는 자신들의 주소 공간에 따로 가져야 하며, 코드의 내용이 실행 도중 변하지 않아야 하므로 재진입 코드(Reentrant or Pure Code)로 컴파일 되어 있다.
        * 페이징에서의 공유는 프로세스 각자의 페이지 테이블에서 엔트리에 같은 프레임 번호를 갖게 함으로써 쉽게 구현 가능하다.


* 2단계 페이징
    * 32 비트를 사용해 주소를 표현하는 시스템에서 하위 12 비트를 offset으로 사용한다면, 페이지의 크기는 2^12 -> 4Kbyte가 되고, Page Table은 최대 2^20 -> 100만개의 엔트리를 가질 수 있게 된다.
        * 이는 4MByte의 값이 나옴.
    * 따라서 페이지 테이블도 작게 나누어 필요한 부분만을 메모리에 적재하는 계층 구조가 도입된다.
    
    * 페이지 번호를 나타내는 20 비트를 10 비트씩 나누어 상위 10비트를 바깥(또는 루트) 페이지 테이블의 엔트리 위치로, 하위 10비트는 나뉘어진 테이블 내의 엔트리 위치를 나타내도록 하면
        * 메모리에 상주시키는 4 Kbyte 크기의 루트 테이블과, 같은 크기의 2^10개 페이지 테이블의 선택적인 메모리 적재를 통해 전부 적재해야 하는 부담을 줄일 수 있다.


* Inverted Page Table
    * 메모리에 고정 크기의 페이지 테이블 하나만 둠으로 페이지 테이블의 큰 사이즈 문제를 해결
    * 메모리의 프레임 수만큼 엔트리를 가지며, 테이블 내에서 엔트리의 순서는 프레임의 순서와 같다.
        * K 번째 엔트리는 메모리의 K 번째 프레임에 대한 사상 정보를 가지며, 존재 비트와 함께 어떤 프로세스의 몇 번째 페이지인지에 대한 정보까지 저장한다.
    
    * 가상 주소 접근을 위해선 pid와 p, d 값을 모두 알아야 한다.
        * 프로세스 번호와 페이지 번호를 함께 mapping 하여 페이지 테이블을 검색.

    * 다만 이 방식은 프레임마다 적재한 프로세스 번호를 가져야 하므로 공유를 구현하기가 힘들다.


* 페이징은 고정 크기의 페이지로 메모리를 관리하기 때문에 구현이 쉽다는 장점으로 대부분의 시스템에서 채택된다.
    * 하지만, 프로그램의 마지막 페이지는 보통 내부 단편화가 발생하고,
    * 서브루틴 등의 논리적인 한 단위가 두 페이지로 나뉘어버린다면 일부가 적재되는 방식에서 mapping이 길어지고 공유 또한 복잡헤질 수 있다.
        * 이런 점 때문에 세그멘테이션이 나옴.


### Segmentation 세그멘테이션
